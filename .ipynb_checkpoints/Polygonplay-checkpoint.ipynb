```python
%matplotlib notebook
import pandas, math, itertools
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Polygon,Point
```

Building the discrete grid

```python
delta = 5
a = np.arange(.5,491,delta)
b = np.arange(.5,680,delta)
c = np.arange(.5,281,delta)
grid = itertools.product(a,b,c)
```

Now load polygons from files

```python
file_names = range(26, 167)
polygons = {}
for f in file_names:
    vertices = pandas.read_csv('parsed_data/sp4/{}.dat'.format(f), names=['x','y','z']).as_matrix()
    polygons[f] = Polygon(vertices)
```

```python
flags = np.zeros((len(a)*len(b)*len(c),1))
plot_points = []
i = 0
for pt in grid:
    z = round(pt[2]) 
    if i%10000 == 0:
        print i,'/', len(a)*len(b)*len(c)
    if z in polygons.keys():
        if polygons[z].contains(Point(pt[0], pt[1], pt[2])):
            flags[i] = 1
            plot_points.append(pt)
    i+=1
```
